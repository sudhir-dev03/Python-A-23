```python
# Tuples are ordered collection of elements 
# Tuples are immutable in nature -- once it is created it cannt be changed after the creation 

# Tuples are iniatialized with paranthesis but there should be one element and one (,)
```

```python
# intializing tuple --
t = (1,2,3,4)
print(type(t))
```

```python
# initiasing a empty tuple
t = ()
print(type(t))
```

```python
# tuple packing which is known like when user give continouos value in group without paranthesis it will automatically saved in tuple
# and known as tuple packing
t = 1,2,3
print(type(t))
```

```python
t = ([1,2],[3,4])
t[0][1] = 100 
print(t)

# here the list is changed not the tuple. Here the whole list is considered as one student 
```

```python
# tuple unpacking -- it allows you you to assign the individual element if tuple to multiple variables in a single line of code 
vars = 1,2,3
a,b,c = vars
# a = vars[0]
# b = vars[1]
# c = vars[2]
print(a,b,c)
```

```python
t = (1,2,2,2,3,4,5)
# since the tuples are immutable so, it has only two mwthods.
#1. count == which is used to find the frequency of the element. 
#1. index == it is used find the index of the element.
t.count(2)
t.index(5)
```

```python
t1 = (1,2,3)
t2 = (4,5,6)
# when we concat two tuples it forms new tuple with the concated tuples
t1+t2
```

```python
# find the some of all the elements of the given tuple 
t = (1,2,3,4,5,6,7,8)
sum =0 
for i in range(len(t)):
    sum += t[i]
print(sum)
```

```python
# convert a tuple of string into one string 
t = ('r','a','h','u','l')
for i in range(len(t)):
    print(t[i],end='')
```
