```python
num = int(input("enter a number:"))
for i in range(1,num+1):
    for j in range(1,i+1):
        print("*",end = " ")
    print()
```

```python
# check wearher a list is sorted or not 
# [4,3,2,1,10] = is not a sorted list 
# [1,2,3,4] = is a sorted list
li = [4,3,2,1,10]
li2 = [1,2,3,4]
for i in range(len(li2)-1):
    if li[i] > li[i+1]:
        print("not sorted")
        break
else:
    print("sorted")
```

```python
# find the k element in a list if the element exist print the index 
li = [1,2,3,4]
k = int(input("enter the element to search:"))
for i in range(len(li)):
    if li[i] == k:
        print("element found at index:",i)
        break
else:
    print("element not found")
```

```python
# write a program to take 4 numbers and take into a list and print the sum of the numbers
li = []
sum = 0
for i in range(4):  
    num = int(input("enter a number:"))
    li.append(num)
    sum += li[i]
print("the list is:",li)
print("the sum of the numbers is:",sum)
```

```python
n = list(map(int,input("enter numbers:").split()))
# print the sum of numbers
s = 0
for i in n:
    s += i      
print("the sum is:",s)  
```
